```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None  
        self.rear = None   
    
    def enqueue(self, data):
        new_node = Node(data)
        if self.rear is None:
            self.front = self.rear = new_node
            return
        self.rear.next = new_node
        self.rear = new_node
    
    def dequeue(self):
        if self.front is None:
            return None
        dequeued_data = self.front.data
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return dequeued_data
    
    def peek(self):
        if self.front is None:
            return None
        return self.front.data
    
    def is_empty(self):
        return self.front is None
    
    def print_queue(self):
        current = self.front
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
    
    def size(self):
        count = 0
        current = self.front
        while current:
            count += 1
            current = current.next
        return count

    def reverse(self):
        prev = None
        current = self.front
        self.rear = self.front
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.front = prev

    def to_list(self):
        result = []
        current = self.front
        while current:
            result.append(current.data)
            current = current.next
        return result


if __name__ == "__main__":
    queue = Queue()
    queue.enqueue(10)
    queue.enqueue(20)
    queue.enqueue(30)
    print("Queue contents:")
    queue.print_queue()

    print("Front element:", queue.peek())
    print("Dequeued element:", queue.dequeue())

    print("Queue after dequeue:")
    queue.print_queue()

    print("Queue size:", queue.size())
    print("Queue as list:", queue.to_list())

    queue.reverse()
    print("Reversed queue:")
    queue.print_queue()
```
